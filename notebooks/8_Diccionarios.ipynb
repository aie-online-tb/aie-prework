#### **8. Diccionarios**

¿Qué es un diccionario?
Un diccionario almacena datos en pares **clave: valor**. En lugar de acceder a los elementos por su posición (como en las listas), accedes por su nombre (la clave). Piensa en él como un formulario: tiene campos con nombre y cada campo tiene un valor.

Los diccionarios son muy habituales en programación porque reflejan bien la realidad: un usuario tiene nombre, email y edad; un producto tiene nombre, precio y stock. En el bootcamp los verás constantemente.

A continuación tienes varios ejemplos. Ejecuta cada celda para ver el resultado:

1. Crear un diccionario

````python
persona = {
    "nombre": "Ana",
    "edad": 28,
    "ciudad": "Madrid"
}

print(persona)

# usamos llaves {} para crear un diccionario
# cada par clave: valor se separa por comas
# las claves suelen ser strings (texto entre comillas)
# los valores pueden ser cualquier tipo: string, número, lista, otro diccionario...
````

2. Acceder a un valor por su clave

````python
persona = {"nombre": "Ana", "edad": 28, "ciudad": "Madrid"}

print(persona["nombre"])  # Ana
print(persona["edad"])    # 28

# ponemos la clave entre corchetes, igual que el índice en las listas
# pero aquí usamos el nombre del campo, no un número
````

3. Error típico: acceder a una clave que no existe

````python
# ESTO DA ERROR: ejecuta la celda para verlo
persona = {"nombre": "Ana", "edad": 28}

print(persona["email"])

# KeyError: 'email'
# el diccionario no tiene ninguna clave llamada 'email'
# las claves distinguen mayúsculas: "Nombre" y "nombre" son claves distintas
````

````python
# Para comprobar si existe una clave antes de acceder:
persona = {"nombre": "Ana", "edad": 28}

if "email" in persona:
    print(persona["email"])
else:
    print("No tiene email registrado")
````

4. Modificar un valor existente

````python
producto = {"nombre": "Portátil", "precio": 999, "stock": 5}

producto["precio"] = 899  # cambiamos el precio

print(producto)

# la misma sintaxis que para acceder, pero con = para asignar
# los diccionarios son modificables, igual que las listas
````

5. Añadir una nueva clave

````python
persona = {"nombre": "Ana", "edad": 28}

persona["email"] = "ana@ejemplo.com"  # nueva clave

print(persona)

# si la clave no existe, Python la crea automáticamente
# si ya existe, la sobreescribe (igual que en el ejemplo anterior)
````

6. Eliminar una clave

````python
persona = {"nombre": "Ana", "edad": 28, "ciudad": "Madrid"}

del persona["ciudad"]

print(persona)

# del elimina la clave y su valor del diccionario
````

7. Obtener todas las claves, valores o pares

````python
persona = {"nombre": "Ana", "edad": 28, "ciudad": "Madrid"}

print(persona.keys())    # dict_keys(['nombre', 'edad', 'ciudad'])
print(persona.values())  # dict_values(['Ana', 28, 'Madrid'])
print(persona.items())   # dict_items([('nombre', 'Ana'), ('edad', 28), ('ciudad', 'Madrid')])

# .keys()   → todas las claves
# .values() → todos los valores
# .items()  → pares (clave, valor). Muy útil para recorrer el diccionario
````

8. Recorrer un diccionario con un bucle

````python
persona = {"nombre": "Ana", "edad": 28, "ciudad": "Madrid"}

for clave, valor in persona.items():
    print(clave, ":", valor)

# .items() devuelve pares (clave, valor)
# el bucle for desempaqueta cada par en dos variables: clave y valor
# esto es muy útil para mostrar o procesar todos los campos de un diccionario
````

9. Diccionario anidado: un diccionario dentro de otro

````python
usuario = {
    "nombre": "Carlos",
    "edad": 35,
    "direccion": {
        "calle": "Gran Vía 42",
        "ciudad": "Barcelona",
        "codigo_postal": "08010"
    }
}

print(usuario["nombre"])                       # Carlos
print(usuario["direccion"]["ciudad"])          # Barcelona
print(usuario["direccion"]["codigo_postal"])   # 08010

# para acceder a un valor anidado, encadenamos los corchetes
# primero accedemos al diccionario interior, luego al valor dentro de él
# esta estructura es MUY común: las APIs de IA devuelven datos en este formato
````

10. Lista de diccionarios: una estructura muy habitual

````python
alumnos = [
    {"nombre": "Ana",    "nota": 8.5},
    {"nombre": "Carlos", "nota": 6.0},
    {"nombre": "Lucía",  "nota": 9.2}
]

for alumno in alumnos:
    print(alumno["nombre"], "→", alumno["nota"])

# una lista de diccionarios = varios registros con los mismos campos
# es como una tabla: cada diccionario es una fila, cada clave es una columna
# verás esta estructura constantemente cuando trabajes con datos
````
