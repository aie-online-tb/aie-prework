### 1. Celdas: los bloques del notebook

Un notebook se compone de **celdas**. Hay dos tipos:

| Tipo | Para qué sirve | Cómo se ve |
|---|---|---|
| **Celda de texto** (Markdown) | Explicaciones, títulos, instrucciones | Texto formateado, como este párrafo |
| **Celda de código** | Código Python que se puede ejecutar | Fondo grisáceo, con un botón ▶ a la izquierda |

La celda de abajo es una **celda de código**. Todavía no ha sido ejecutada.

```python
# Esto es una celda de código
# Las líneas que empiezan con # son comentarios. Python las ignora
# Son notas para el humano que lee el código

print("¡Hola! Esta es mi primera celda de código.")
```

### 2. Añadir, mover y eliminar celdas

Puedes gestionar las celdas del notebook de las siguientes formas:

- **Añadir una celda nueva:**  
Haz clic en **+ Código** o **+ Texto** en la barra superior, o pasa el ratón entre dos celdas y aparecerán los mismos botones.

- **Eliminar una celda:**  
Haz clic en los tres puntos (`...`) que aparecen a la derecha de la celda y selecciona **Eliminar celda**. También puedes hacer clic en la celda y pulsar dos veces la tecla `D`.

- **Mover una celda:**  
Usa los botones de flecha arriba y abajo que aparecen al pasar el ratón sobre la celda, o arrástrala desde el margen izquierdo.

- **Ejecutar todas las celdas de golpe:**  
Ve a **Entorno de ejecución > Ejecutar todo** (`Ctrl + F9`). Útil para relanzar el notebook tras un reinicio.

Durante el prework no necesitarás añadir ni eliminar celdas. Pero conviene que sepas que puedes hacerlo.

### 3. Cómo ejecutar una celda de código

Para ejecutar una celda de código tienes dos opciones:

1. **Haz clic en el botón ▶** que aparece a la izquierda de la celda
2. **Pulsa `Shift + Enter`** con el cursor dentro de la celda (atajo recomendado)

Cuando una celda se ejecuta:
- Aparece un número entre corchetes a la izquierda: `[1]`. Ese número indica el orden en que fue ejecutada.
- El **resultado** (output) aparece justo debajo de la celda.

**Pruébalo ahora:** ejecuta la celda de código del apartado anterior y mira qué aparece debajo.

### 4. El orden de ejecución importa

Las celdas de un notebook **comparten memoria**. Si defines una variable en una celda, puedes usarla en las siguientes. Pero para que eso funcione, **debes ejecutar las celdas en orden, de arriba a abajo**.

Prueba a ejecutar las dos celdas siguientes en orden:

```python
# Celda A: ejecuta esta primero
nombre = "Ana"
print("Variable creada:", nombre)
```

```python
# Celda B: ejecuta esta después
# Si ejecutas esta celda SIN haber ejecutado la celda A primero, dará error
# porque la variable nombre todavía no existe
print("Hola,", nombre)
```

**Regla de oro:** al empezar a trabajar en un notebook, ejecuta siempre las celdas **de arriba a abajo**, sin saltarte ninguna.

En Google Colab puedes ejecutar todas las celdas de golpe desde el menú: **Entorno de ejecución → Ejecutar todo**.

### 5. Qué hacer cuando aparece un error

Los errores son normales, todos los programadores los ven constantemente. Un error no significa que hayas roto nada.

Cuando una celda falla, Python muestra un mensaje de error en rojo debajo de la celda.  

Lo importante es **leer el mensaje**: casi siempre te dice exactamente qué ha fallado y en qué línea.

Ejecuta la celda de abajo para ver un error a propósito:

```python
# ESTO DA ERROR: ejecútalo para ver cómo se ve
print(variable_que_no_existe)

# NameError: name 'variable_que_no_existe' is not defined
# Python te dice: intentaste usar una variable que no has creado todavía
# Lee siempre la última línea del error. Suele ser la más informativa
```

A lo largo del prework encontrarás celdas marcadas con `# ESTO DA ERROR`. Son **intencionales**: las usamos para mostrarte cómo se ve un error y qué significa. Ejecútalas sin miedo, no pasa nada.

### 6. El entorno se reinicia si cierras el navegador

Cuando cierras Google Colab o pasa un tiempo sin actividad, el **entorno de ejecución se reinicia**. Esto significa que todas las variables que habías creado desaparecen.

**No pasa nada**, el código sigue ahí. Solo tienes que volver a ejecutar las celdas desde el principio.

Si al abrir un notebook ves este mensaje de error al ejecutar una celda:

> `NameError: name 'X' is not defined`

...probablemente el motivo es que el entorno se reinició. Solución: **Entorno de ejecución → Ejecutar todo**.
