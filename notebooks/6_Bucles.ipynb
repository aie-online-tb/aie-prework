#### **6. Bucles**

¿Qué es un bucle?
Un bucle es una instrucción que le dice a Python: *"repite esto varias veces"*. Sin bucles, si quisieras imprimir 100 números tendrías que escribir 100 líneas de código. Con un bucle, son 2 líneas.

Hay dos tipos principales de bucles en Python: `for` y `while`. El más usado con diferencia es `for`.

A continuación tienes varios ejemplos. Ejecuta cada celda para ver el resultado:

1. Bucle `for` sobre una lista

```python
frutas = ["manzana", "naranja", "kiwi"]

for fruta in frutas:
    print(fruta)

# el bucle for toma cada elemento de la lista frutas
# le pone el nombre fruta (lo inventamos nosotros, podría llamarse x, f, elemento...)
# y ejecuta el bloque indentado (el print) para cada uno
# este bucle da 3 vueltas porque la lista tiene 3 elementos
```

2. Bucle `for` con `range()`: contar con un bucle

```python
for numero in range(5):
    print(numero)

# range(5) genera los números 0, 1, 2, 3, 4
# ¡empieza en 0 y el 5 NO está incluido!
# es muy útil cuando quieres repetir algo un número exacto de veces
```

3. `range()` con inicio y fin

```python
for numero in range(1, 6):
    print(numero)

# range(inicio, fin) genera los números desde inicio hasta fin-1
# range(1, 6) genera: 1, 2, 3, 4, 5
# el fin nunca está incluido, sigue la misma lógica que los índices de las listas
```

4. Hacer algo con cada elemento en el bucle

```python
precios = [10, 25, 8, 40, 15]
total = 0

for precio in precios:
    total = total + precio

print("Total:", total)

# en cada vuelta del bucle tomamos el precio y lo sumamos al total
# total empieza en 0 y va acumulando
# al final del bucle, total contiene la suma de todos los precios
```

5. Bucle `while`: repite mientras una condición sea verdadera

```python
contador = 0

while contador < 5:
    print(contador)
    contador = contador + 1

# mientras contador sea menor que 5, ejecuta el bloque
# en cada vuelta imprimimos el contador y lo incrementamos en 1
# cuando contador llega a 5, la condición es False y el bucle para
# ¡si te olvidas de incrementar el contador, el bucle nunca para!
```

6. `break`: cómo salir del bucle antes de tiempo

```python
numeros = [3, 7, 2, 9, 1, 5]

for numero in numeros:
    if numero == 9:
        print("¡Encontré el 9! Paro el bucle.")
        break
    print(numero)

# cuando numero == 9: imprime el mensaje y hace break (para el bucle)
# cuando numero != 9: salta el if y llega al print(numero), por eso 3, 7 y 2 SÍ se imprimen
# break es útil cuando ya encontraste lo que buscabas y no tiene sentido seguir
```

7. `continue`: cómo saltar una vuelta y seguir

```python
numeros = [1, 2, 3, 4, 5, 6]

for numero in numeros:
    if numero % 2 == 0:
        continue
    print(numero)

# continue salta el resto del bloque y pasa a la siguiente vuelta
# aquí: si el número es par (resto 0 al dividir por 2), lo saltamos
# solo se imprimen los impares: 1, 3, 5
```

8. Error típico: olvidar la indentación

```python
# ESTO DA ERROR: ejecuta la celda para verlo
for i in range(3):
print(i)

# IndentationError: expected an indented block
# Python usa la indentación (sangría) para saber qué está DENTRO del bucle
# todo lo que va dentro del for debe tener 4 espacios (o 1 tab) de margen
```

```python
# ASÍ ES CORRECTO
for i in range(3):
    print(i)  # 4 espacios de indentación
```

9. Bucle `for` con strings (texto)

```python
nombre = "Python"

for letra in nombre:
    print(letra)

# Python también puede iterar sobre un string letra por letra
# el for funciona con cualquier cosa que sea iterable (listas, strings, rangos...)
```

10. Bucles anidados: un bucle dentro de otro

```python
colores = ["rojo", "azul"]
tallas = ["S", "M", "L"]

for color in colores:
    for talla in tallas:
        print(color, talla)

# el bucle exterior da 2 vueltas (una por color)
# en cada vuelta, el interior da 3 vueltas (una por talla)
# total: 6 combinaciones
```

##### Solución Problemas Propuestos

**Problema 1:**

Crea una lista con los nombres de 5 países y usa un bucle `for` para imprimirlos todos.

```python
paises = ["España", "México", "Argentina", "Colombia", "Chile"]

for pais in paises:
    print(pais)
```

**Problema 2:**

Usa `range()` para imprimir los números del 1 al 10.

```python
for numero in range(1, 11):
    print(numero)
```

**Problema 3:**

Dada la lista `notas = [7, 4, 9, 6, 3, 8]`, usa un bucle para imprimir solo las notas que sean 5 o más.

```python
notas = [7, 4, 9, 6, 3, 8]

for nota in notas:
    if nota >= 5:
        print(nota)
```

**Problema 4:**

Usando un bucle `while`, imprime los números del 10 al 1 (en orden descendente).

```python
contador = 10

while contador >= 1:
    print(contador)
    contador = contador - 1
```
