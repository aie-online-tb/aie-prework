#### **7. Funciones**

¿Qué es una función?
Una función es un bloque de código al que le pones un nombre, y que puedes ejecutar cuando quieras simplemente llamando a ese nombre. Las funciones evitan repetir el mismo código una y otra vez.

Ya has usado funciones antes sin saberlo: `print()`, `len()`, `input()` son todas funciones. Ahora aprenderás a crear las tuyas.

A continuación tienes varios ejemplos. Ejecuta cada celda para ver el resultado:

1. Función básica sin parámetros

```python
def saludar():
    print("¡Hola! Bienvenido al bootcamp.")

saludar()

# def = define (le dices a Python que vas a crear una función)
# saludar = el nombre que le damos (lo inventamos nosotros)
# () = los paréntesis son obligatorios
# : = los dos puntos indican que empieza el bloque de la función
# el código indentado es el cuerpo de la función
# saludar() = aquí es donde LLAMAMOS (ejecutamos) la función
```

2. Error típico: llamar a la función antes de definirla

```python
# ESTO DA ERROR: ejecuta la celda para verlo
despedirse()

def despedirse():
    print("¡Hasta luego!")

# NameError: name 'despedirse' is not defined
# Python lee el código de arriba a abajo
# cuando intenta ejecutar despedirse(), la función todavía no existe
# siempre define la función ANTES de llamarla
```

```python
# ASÍ ES CORRECTO: primero defines, luego llamas
def despedirse():
    print("¡Hasta luego!")

despedirse()
```

3. Función con parámetros

```python
def saludar_persona(nombre):
    print("¡Hola,", nombre + "!")

saludar_persona("Ana")
saludar_persona("Carlos")
saludar_persona("Lucía")

# nombre es un parámetro: una variable que recibe la función al ser llamada
# cada vez que llamamos a la función le pasamos un valor diferente
# la misma función, tres resultados distintos. Eso es la potencia de las funciones
```

4. Función con varios parámetros

```python
def presentar(nombre, edad):
    print(nombre, "tiene", edad, "años")

presentar("Ana", 28)
presentar("Carlos", 35)

# el orden importa: el primer valor va al primer parámetro, el segundo al segundo
# presentar("Ana", 28) → nombre = "Ana", edad = 28
```

5. Función con `return`: cómo devolver un valor

```python
def sumar(a, b):
    return a + b

resultado = sumar(3, 4)
print(resultado)  # 7

print(sumar(10, 20))  # 30

# return devuelve un valor al lugar donde se llamó la función
# ese valor podemos guardarlo en una variable o usarlo directamente
# sin return, la función hace cosas pero no produce ningún valor
```

6. Error típico: olvidar el `return`

```python
def multiplicar(a, b):
    a * b  # ¡falta el return!

resultado = multiplicar(4, 5)
print(resultado)

# imprime: None
# la función calculó 4*5 pero no devolvió nada
# cuando una función no tiene return, Python devuelve None automáticamente
```

```python
# ASÍ ES CORRECTO
def multiplicar(a, b):
    return a * b

resultado = multiplicar(4, 5)
print(resultado)  # 20
```

7. Parámetros con valor por defecto

```python
def calcular_precio(precio, descuento=0):
    return precio - descuento

print(calcular_precio(100))       # sin descuento → 100
print(calcular_precio(100, 15))   # con descuento → 85

# descuento=0 es el valor por defecto
# si llamamos la función sin ese argumento, usa 0
# si le pasamos un valor, usa ese valor
# los parámetros con valor por defecto siempre van al final
```

8. Scope: las variables definidas dentro no existen fuera

```python
# ESTO DA ERROR: ejecuta la celda para verlo
def calcular_area(base, altura):
    area = base * altura
    return area

calcular_area(5, 3)
print(area)  # NameError: name 'area' is not defined

# area existe DENTRO de la función, pero no fuera
# cada función tiene su propio espacio (scope)
# para usar el resultado fuera, hay que guardarlo con return
```

```python
# ASÍ ES CORRECTO: guardamos el resultado en una variable externa
def calcular_area(base, altura):
    area = base * altura
    return area

mi_area = calcular_area(5, 3)
print(mi_area)  # 15
```

9. Función que usa otras funciones

```python
def al_cuadrado(numero):
    return numero * numero

def suma_de_cuadrados(a, b):
    return al_cuadrado(a) + al_cuadrado(b)

print(suma_de_cuadrados(3, 4))  # 9 + 16 = 25

# las funciones pueden llamar a otras funciones
# esto permite descomponer problemas complejos en piezas pequeñas
```

10. Reutilización: la clave de las funciones

```python
def calcular_iva(precio, porcentaje=21):
    return precio * porcentaje / 100

productos = ["Portátil", "Ratón", "Teclado"]
precios =   [999,         25,       79]

for i in range(len(productos)):
    iva = calcular_iva(precios[i])
    print(productos[i], "→ IVA:", iva, "€")

# combinamos todo lo aprendido: función + bucle + listas
# la función calcular_iva se usa 3 veces sin repetir código
```

##### Solución Problemas Propuestos

**Problema 1:**

```python
def saludar_con_nombre(nombre):
    print("Hola,", nombre + "! ¿Cómo estás?")

saludar_con_nombre("Ana")
saludar_con_nombre("Carlos")
```

**Problema 2:**

```python
def es_mayor_de_edad(edad):
    return edad >= 18

print(es_mayor_de_edad(20))  # True
print(es_mayor_de_edad(15))  # False
```

**Problema 3:**

```python
def calcular_media(numeros):
    total = 0
    for numero in numeros:
        total = total + numero
    return total / len(numeros)

print(calcular_media([5, 7, 3, 9, 6]))  # 6.0
```

**Problema 4:**

```python
def contar_aprobados(notas):
    aprobados = 0
    for nota in notas:
        if nota >= 5:
            aprobados = aprobados + 1
    return aprobados

notas = [7, 4, 9, 6, 3, 8, 2, 5]
print(contar_aprobados(notas))  # 5
```
